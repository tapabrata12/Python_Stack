## 4. Sep Parameter

- `read_csv()`: Usually takes comma separated value but if we have tab separated value then we can change the `sep` parameter to `'/t'` for reading `.tsv` files

```python
df3 = pd.read_csv(str(dir)+'/Datasets/movie_titles_metadata.tsv')
```

```python
df3
```

```python
df3 = pd.read_csv(str(dir) + '/Datasets/movie_titles_metadata.tsv', sep='\t')
```

```python
df3
```

-  If you see you can understand that columns are not properly shown, these kind of probleam offen happen in `.tsv` files. So have to give names by our own using `names` attribute

```python
df3 = pd.read_csv(str(dir) + '/Datasets/movie_titles_metadata.tsv', sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

```python
df3
```

### 14. Convertors

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
df12 = pd.read_csv(str(dir)+'/Datasets/IPL Matches 2008-2020.csv')
```

```python
df12
```

```python
df12 = pd.read_csv(str(dir)+'/Datasets/IPL Matches 2008-2020.csv',converters={'team1':rename})
```

```python
df12
```

- Undert team 1 column all'Royal Challengers Bangalore' converted to'RCB'
