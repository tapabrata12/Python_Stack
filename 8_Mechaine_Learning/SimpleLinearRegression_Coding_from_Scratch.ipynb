### **Code Implementation**

```python
class SimpleLinearRegression:
  def __init__(self):
    """
    Defining Coff and Intercept as None at first
    """
    self.m = None
    self.b = None

  """
  Making fit method to train the model
  """
  def fit(self,X_train,y_train):
    numarator = 0
    denominator = 0
    for i in range(len(X_train)):
      numarator = numarator + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
      denominator = denominator + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))


    self.m = numarator/denominator
    self.b = y_train.mean() - (self.m * X_train.mean())
    # print(self.m)
    # print(self.b)
    return


  """
  Making predict method to predict the values
  """
  def predict(self, X_test): # Added self as the first argument

    return self.m * X_test + self.b
```
