# Decision Boundary Drown By the SVM Model

```python
# Prepare features and labels
from sklearn import svm
X_data = data.iloc[:, :2].values  # Use first two columns as features
y_data = data['target'].values    # Use 'target' column as labels

# Create and fit the SVM model
svm_model = svm.SVC(kernel='linear')
svm_model.fit(X_data, y_data)

# Plot the data points
plt.scatter(X_data[:, 0], X_data[:, 1], c=y_data, s=30, cmap=plt.cm.prism)

# Create the axis bounds for the grid
axis = plt.gca()
x_limit = axis.get_xlim()
y_limit = axis.get_ylim()

# Create a grid to evaluate the model
x = np.linspace(x_limit[0], x_limit[1], 50)
y = np.linspace(y_limit[0], y_limit[1], 50)
X, Y = np.meshgrid(x, y)
xy = np.c_[X.ravel(), Y.ravel()]

# Get the decision function
decision_line = svm_model.decision_function(xy).reshape(Y.shape)

# Plot the decision boundary
axis.contour(X, Y, decision_line, colors='k', levels=[0], linestyles=['-'])

# Plot support vectors
axis.scatter(svm_model.support_vectors_[:, 0], svm_model.support_vectors_[:, 1],
             s=100, linewidth=1, facecolors='none', edgecolors='k')

# Show plot
plt.show()
```
