```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path # I include this so that it will run to all the places with giving error
directory = Path(__name__).parent.absolute()
```

```python
data = pd.read_csv(str(directory)+'/Datasets/Mall_Customers.csv')
```

```python
data.head()
```

```python
plt.scatter(x='Annual Income (k$)', y='Spending Score (1-100)',data=data)
```

```python
X = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
km.fit(X)
```

```python
y = km.predict(X)
```

```python
y
```

```python
X['Cluster'] = y
```

```python
X.head()
```

```python
sns.scatterplot(data=X, x='Annual Income (k$)', y='Spending Score (1-100)',hue='Cluster',palette=['red','green','blue','yellow','black'])
```
