```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path # I include this so that it will run to all the places with giving error
directory = Path(__name__).parent.absolute()
```

```python
data = pd.read_csv(str(directory)+'/Datasets/IRIS.csv')
```

```python
data.head()
```

```python
X = data.drop('species',axis=1)
```

```python
X.head()
```

```python
y = data['species']
```

```python
y.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X,y)
```

```python
knn.score(X,knn.predict(X))
```
