```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Basic Array Plot

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = x*2
y
```

### Using matplotlib with `plt.plot` function

```python
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Array Plot")
plt.show()
```

- Adding some extra things

```python
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Array Plot")
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

### Exporting the plot

```python
from pathlib import Path # I include this so that it will run to all the places with giving error
directory = Path(__name__).parent.absolute()
plt.plot(x,y)
plt.savefig(str(directory)+'/Plots/Plot1')
```

- `plt.subplots()`

```python
fig, axes = plt.subplots()

axes.plot(x,y,'y')
axes.set_xlabel('X')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes.shape
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for i in axes:
    i.plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('0 0 Title')
axes[0][1].plot(y,x)
axes[0][1].set_title('0 1 Title')

axes[1][0].plot(b,a)
axes[1][0].set_title('1 0 Title')
axes[1][1].plot(a,b)
axes[1][1].set_title('1 1 Title')


plt.tight_layout()
plt.savefig(str(directory)+'/Plots/Subplots')
```
