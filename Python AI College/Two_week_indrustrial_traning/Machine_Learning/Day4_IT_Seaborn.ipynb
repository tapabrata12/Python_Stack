# Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path # I include this so that it will run to all the places with giving error
directory = Path(__name__).parent.absolute()
```

```python
data = pd.read_csv(str(directory)+'/Datasets/dm_office_sales.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

## Scatterplot

```python
sns.scatterplot(x='salary', y='sales' , data=data)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' , data=data)
```

## Seaborn Parameters

- `Hue`

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' , data=data, hue='division')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' , data=data, hue='work experience',palette=['Crimson', 'Blue','Green','Goldenrod',
'Coral',
'Lavender',           
'Turquoise',
'Maroon',
'Teal',
'red',
'Indigo',
'yellow',
'black',
'Pink',
])
```

- `size`

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' , data=data, size='work experience')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' , data=data, s=200)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary', y='sales' ,style='division', data=data, hue='division')
plt.savefig(str(directory)+'/Plots/Example')
```
