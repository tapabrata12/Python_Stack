# Decision Tree 

```python
from pathlib import Path
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X_Path = Path(__name__).parent.absolute()
# Load dataset
df = pd.read_csv(str(X_Path)+"/datasets/data_with_department.csv")

# Convert categorical values into numeric codes
df['Nationality'] = df['Nationality'].map({'UK': 0, 'USA': 1, 'N': 2})
df['Department'] = df['Department'].map({'HR': 0, 'IT': 1, 'Finance': 2})
df['Go'] = df['Go'].map({'YES': 1, 'NO': 0})

#  Define input features and target
features = ['Age', 'Experience', 'Rank', 'Nationality', 'Department']
X = df[features]
y = df['Go']

# Create and train the Decision Tree
dtree = DecisionTreeClassifier(criterion='entropy')  # or use 'gini'
dtree.fit(X, y)

#  Visualize the Decision Tree
plt.figure(figsize=(14, 8))
plot_tree(dtree, 
          feature_names=features, 
          class_names=['NO', 'YES'], 
          filled=True, 
          rounded=True)
plt.title("Decision Tree with 'Department' Feature")
plt.show()
```
