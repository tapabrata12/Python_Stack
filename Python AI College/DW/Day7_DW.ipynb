## Data Visulization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample data of 50 students
np.random.seed(10)
data = {
    'Math': np.random.randint(40, 100, 50),
    'Science': np.random.randint(35, 95, 50),
    'English': np.random.randint(30, 90, 50),
    'History': np.random.randint(40, 85, 50)
}

df = pd.DataFrame(data)
print(df)

# Histogram of Math Marks
plt.hist(df['Math'], bins=8, color='skyblue', edgecolor='black')
plt.title('Histogram of Math Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.show()

# Boxplot of all subjects
plt.figure(figsize=(8,5))
sns.boxplot(data=df, palette='Set3')
plt.title('Boxplot of Marks in All Subjects')
plt.ylabel('Marks')
plt.show()

# Scatter plot between Math and Science
plt.scatter(df['Math'], df['Science'], color='tomato')
plt.title('Math vs Science Marks')
plt.xlabel('Math Marks')
plt.ylabel('Science Marks')
plt.show()

# Line plot of average marks per subject
avg_marks = df.mean()
avg_marks.plot(kind='line', marker='o', color='indigo')
plt.title('Average Marks in Each Subject')
plt.ylabel('Average Marks')
plt.grid(True)
plt.show()
```
