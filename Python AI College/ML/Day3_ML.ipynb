# Plot the different results obtained by algebraic and statistical operations in python

- ### (i) Basic algebraic operations: addition, subtraction, multiplication, division
- ### (ii) Statistical operations: mean, median, standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Data
x = np.linspace(1, 10, 10)  # 10 values from 1 to 10
y1 = np.random.randint(1, 20, 10)  # Random values between 1 and 20
y2 = np.random.randint(1, 20, 10)  # Another set of random values

# Step 2: Perform Algebraic Operations
addition_result = y1 + y2
subtraction_result = y1 - y2
multiplication_result = y1 * y2
division_result = np.divide(y1, y2, where=y2 != 0)  # Avoid division by zero

# Step 3: Compute Statistical Operations for y1
mean_y1 = np.mean(y1)
median_y1 = np.median(y1)
std_y1 = np.std(y1)

# Step 4: Create a Figure for Plotting
plt.figure(figsize=(15, 10))

# Plot Addition
plt.subplot(2, 3, 1)
plt.plot(x, y1, 'b-o', label="y1")
plt.plot(x, y2, 'r-o', label="y2")
plt.plot(x, addition_result, 'g-o', label="y1 + y2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Addition Operation")
plt.legend()
plt.grid()

# Plot Subtraction
plt.subplot(2, 3, 2)
plt.plot(x, y1, 'b-o', label="y1")
plt.plot(x, y2, 'r-o', label="y2")
plt.plot(x, subtraction_result, 'g-o', label="y1 - y2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Subtraction Operation")
plt.legend()
plt.grid()

# Plot Multiplication
plt.subplot(2, 3, 3)
plt.plot(x, y1, 'b-o', label="y1")
plt.plot(x, y2, 'r-o', label="y2")
plt.plot(x, multiplication_result, 'g-o', label="y1 * y2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Multiplication Operation")
plt.legend()
plt.grid()

# Plot Division
plt.subplot(2, 3, 4)
plt.plot(x, y1, 'b-o', label="y1")
plt.plot(x, y2, 'r-o', label="y2")
plt.plot(x, division_result, 'g-o', label="y1 / y2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Division Operation")
plt.legend()
plt.grid()

# Plot Statistical Operations
plt.subplot(2, 3, 5)
plt.plot(x, y1, 'b-o', label="y1")  # Original y1 data
plt.axhline(mean_y1, color='g', linestyle='--', label="Mean")  # Mean line
plt.axhline(median_y1, color='r', linestyle='-.', label="Median")  # Median line
plt.axhline(mean_y1 + std_y1, color='purple', linestyle=':', label="Mean + Std Dev")  # Std Dev upper bound
plt.axhline(mean_y1 - std_y1, color='purple', linestyle=':', label="Mean - Std Dev")  # Std Dev lower bound
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Statistical Operations")
plt.legend()
plt.grid()

# Step 5: Display the Plots
plt.tight_layout()
plt.show()
```
