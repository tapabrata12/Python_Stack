# Python code univarient analysis to implement EDA help of histogram, dataset in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset from seaborn
df = sns.load_dataset('tips')  

# Display first few rows
print(df.head())

# Univariate analysis: Histogram for numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plotting histograms
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

# Python code snippet to perform bivariate analysis using histograms as part of Exploratory Data Analysis (EDA), using Seaborn on a sample dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset as a dictionary-like structure
tips = sns.load_dataset('tips')

# Convert columns to lists (not using pandas operations)
total_bill = tips['total_bill'].to_list()
tip = tips['tip'].to_list()
sex = tips['sex'].to_list()
day = tips['day'].to_list()

# Bivariate Histogram: total_bill with hue = sex
plt.figure(figsize=(10, 6))
sns.histplot(x=total_bill, hue=sex, bins=20, kde=True, element='step', stat='density')
plt.title('Bivariate Histogram: Total Bill by Sex')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()

# Bivariate Histogram: tip with hue = day
plt.figure(figsize=(10, 6))
sns.histplot(x=tip, hue=day, bins=20, multiple='stack')
plt.title('Bivariate Histogram: Tip by Day')
plt.xlabel('Tip')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
